# **El ciclo $while$ en Python**

# Ciclos o bucles.

Los <b>ciclos</b> o <b>bucles</b> permiten ejecutar una o más instrucciones (secuenciales), cero o más veces, con base en el valor de verdad que arroje la evaluación de una expresión de tipo lógico. Esta expresión le permite al algoritmo tomar la decisión de repetir o dejar de ejecutar el grupo de instrucciones.

Python tiene implementado dos estructuras de control cíclicas o de bucle, que son <b><i>while</i></b> y <b><i>for</i></b>. Antes de proceder a verlas es de importancia reconocer la diferencia entre os tipos de variables, las que sirven como contadores y aquellas que se utilizan como acumuladores.

#### Variable: $Contador$

Muchas veces en los procesos repetitivos es necesario hacer el conteo de sucesos o acciones internas del ciclo; este proceso de conteo se hace con una variable que se va incrementando cada vez que el ciclo se repite.

El <b><i>contador</i></b> es una variable que incrementa o disminuye su valor en forma constante cada que ocurre una acción o suceso.

La forma general del <b><i>contador</i></b> es:

$CONTADOR\ =\ CONTADOR\ +\ <valor\ constante> $

```python
cont = 0
cont = cont + 1 # cont = 0 + 1 = 1
cont = cont + 1 # cont = 1 + 1 = 2
cont = cont + 1 # cont = 2 + 1 = 3
cont = cont + 1 # cont = 3 + 1 = 4
print("Cont = ",cont)
```

#### Variable: $Acumulador$

Un <b><i>acumulador</i></b> o totalizador es una variable cuya misión es almacenar cantidades variables resultantes de procesos sucesivos.

La diferencia con el contador radica en que el incremento o disminución es <b><i>variable</i></b> y NO constante.

La forma general del <b><i>acumulador</i></b> es:

$ACUMULADOR\ =\ ACUMULADOR\ +\ <expresión> $

```python
acum = 0
acum = acum + 4.5
acum = acum + 2.9
acum = acum + 3.6
acum = acum + 4.7
print("Acum = ",acum)
print("Promedio de notas = ",acum/4)
```

## Ciclo $while$

Al ejecutarse esta instrucción, la condición es evaluada. Si la condición resulta verdadera ($True$), se ejecuta una vez la secuencia de instrucciones que forman el cuerpo del ciclo. Al finalizar la ejecución del cuerpo del ciclo se vuelve a evaluar la condición y si es verdadera la ejecución se repite, en caso contrario ($False$) se termina la ejecución del cuerpo del ciclo.



while.png



La palabra reservada en Python es <b><i>while</i></b>, luego la condición a evaluar, seguido se usan los dos puntos ( <b><i>:</i></b> ) y para delimitar el código que está incluido en el ciclo se usa la <b><i>indentación</i></b>.



while_text.png

```python
# Inicializando el contador
i = 1

# Condición de parada i<=5
while i <= 5:
    print(i)
    #Incremento del contador
    #i = i + 1
    i += 1

print("\nFin de programa, i = " + str(i))
```

Como se vio en el ejemplo anterior, el ciclo <b><i>while</i></b> se repite un número determinado de veces, porque la condición así lo establece. Pero, tambien es posible crear <b>ciclos</b> o <b>bubles infinitos</b>.

Una forma de entrear en un bucle infinito es no actualizando el contador del ciclo.

```python
# Inicialización del contador
i = 1

# condición de parada
while i <= 5:
    print('Entró a un ciclo infinito - ' , i)

print("\nFin de programa")
```

```python
# Dando al while una condición que siempre sea cierta
while 1==1:
    print('Entró a un ciclo infinito')
```

Como se ve en el ejemplo anterior, no se usan variables en la estructura del <b><i>while</i></b>, todos son contantes, son elementos que no varían en el tiempo.

En la condición del ciclo <b><i>while</i></b> se pueden usar otros tipos datos o variables como <b><i>cadenas</i></b>  y <b><i>booleanos</i></b>

```python
# Usando datos tipos lógicos o booleanos
m = 0
while True:
    print('Entró a un ciclo infinito - ' , m)
    m += 1

print('\nFin de programa.')
```

## Secuencias de control: la sentencia $break$ y $continue$ 

* Para finalizar un bucle while prematuramente, se puede utilizar la sentencia <b><i>break</i></b>.

* Para volver a la primera línea del ciclo debido a una condición dada, se usa el <b><i>continue</i></b>.

```python
# inicalización del contador
i = 0

# contador - incremento constante de la variable
while i<=10:
    i += 1
    if i == 5:
        continue
    else:
        print(i)
```

```python
# inicalización del contador
i = 0

# contador - incremento constante de la variable
while i<=10:
    i += 1
    if i == 5:
        break
    else:
        print(i)
```
